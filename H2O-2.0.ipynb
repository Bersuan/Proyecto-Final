```python
import pandas as pd
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
df_final = pd.read_csv('output/dfFinal.csv')
```

```python
df_final
```

```python
y = df_final.HOME_TEAM_WINS
```

```python
X = df_final.drop(columns =['Unnamed: 0','HOME_TEAM_WINS','SEASON','GAME_ID'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train = pd.concat([X_train, y_train], axis=1, sort=False)
```

```python
test = pd.concat([X_test, y_test], axis=1, sort=False)
```

```python
test
```

```python
h2train = h2o.H2OFrame(train)
h2test = h2o.H2OFrame(test)
```

```python
columnas = [a for a in h2train.columns if a != 'HOME_TEAM_WINS']
x = columnas
y = 'HOME_TEAM_WINS'
print(columnas)
```

```python
automl = H2OAutoML(max_models=50, seed=42,max_runtime_secs=300, sort_metric='RMSE')
automl.train(x=x, y=y, training_frame=h2train)
```

```python
print('[INFO] Models leader board:')
leader_board = automl.leaderboard
leader_board.head()
```

```python
metrics = automl.leader.cross_validation_metrics_summary().as_data_frame()
metrics
```

```python
automl.leader.varimp(use_pandas=True)
```

```python
predictions = automl.leader.predict(h2test)
```

```python
h2o.save_model(automl.leader, path='./mymodel')
```
