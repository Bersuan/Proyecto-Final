```python
import pandas as pd
from sklearn.metrics import *
```

```python
games = pd.read_csv('input/games.csv')
```

```python
ranking = pd.read_csv('input/ranking.csv')
```

TODOS LOS EQUIPOS DE LOCAL

```python
todos = games[games['SEASON']>=2019]
```

```python
df_todos = todos[['HOME_TEAM_ID','PTS_home', 'FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home','AST_home', 'REB_home']]
```

```python
mediasLocales2019 = df_todos.groupby('HOME_TEAM_ID').mean()
mediasLocales2019.loc[1610612764]
```

TODOS LOS EQUIPOS DE VISITANTE

```python
todosVisitantes = games[games['SEASON']>=2019]
```

```python
df_todosVisitantes = todosVisitantes[['VISITOR_TEAM_ID','PTS_away', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away','AST_away', 'REB_away']]
```

```python
mediasVisitantes2019 = df_todosVisitantes.groupby('VISITOR_TEAM_ID').mean()
```

UNIMOS Y LIMPIAMOS DATAFRAMES

```python
df_2019 = games[['HOME_TEAM_ID', 'VISITOR_TEAM_ID','SEASON','HOME_TEAM_WINS']]
```

```python
df_2019 = df_2019[df_2019['SEASON']>=2019]
df_2019
```

```python
df_2019 = df_2019.merge(right=mediasLocales2019, on=['HOME_TEAM_ID'])
df_2019.head()
```

```python
df_2019 = df_2019.merge(right=mediasVisitantes2019, on=['VISITOR_TEAM_ID'])
df_2019.head()
```

```python
df_2019['HOME_TEAM_WINS']= df_2019['HOME_TEAM_WINS'].replace({0: 'VISITANTE', 1: 'LOCAL'})
```

```python
categoricalColumns = ['HOME_TEAM_ID', 'VISITOR_TEAM_ID']
for i in categoricalColumns:
    df_2019[i] = df_2019[i].astype('category')
```

```python
df_2019.head()
```

```python
df_2019.HOME_TEAM_WINS.value_counts()
```

```python
df_2019.dtypes
```

ENTRENAMOS EL MODELO

```python
y = df_2019.HOME_TEAM_WINS
```

```python
X = df_2019.drop(columns =['HOME_TEAM_WINS','SEASON'])
```

```python
print(y.shape)
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=800, max_depth=20, min_samples_leaf=3, random_state=111, n_jobs = -1)

model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
print('Accurancy', accuracy_score(y_test, y_pred))
print("Precision", precision_score(y_test, y_pred, average='weighted'))
print("Recall", recall_score(y_test, y_pred, average='weighted'))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

y_score = model.predict_proba(X_test)


display(pd.DataFrame({
    "real":y_test,
    "predicted":y_pred,
    "probaLocal":y_score[:,0],
    "probaVisitante":y_score[:,1]
    
}).head(30))
```

```python
y_predAll=model.predict(X)
print('Accurancy', accuracy_score(y, y_predAll))
print("Precision", precision_score(y, y_predAll, average='weighted'))
print("Recall", recall_score(y, y_predAll, average='weighted'))
```

```python
teams = pd.read_csv('input/teams.csv')
```

```python
teams = teams[['TEAM_ID', 'ABBREVIATION' ]]
```

```python
teams
```

```python
teams = teams.set_index('ABBREVIATION')
```

```python
def homeTeam (name):
    return teams['TEAM_ID'].loc[name]
```

```python
def awayTeam (name):
    return teams['TEAM_ID'].loc[name]
```

```python
homeTeam('DAL')
```

```python
awayTeam('NOP')
```

```python
def prediceme (local, visitante):
    local = pd.DataFrame(mediasLocales2019.loc[homeTeam(local)])
    visitante = pd.DataFrame(mediasVisitantes2019.loc[awayTeam(visitante)])
    local = local.T
    local.reset_index(inplace=True)
    local = local.rename(columns={'index':'HOME_TEAM_ID'})
    visitante = visitante.T
    visitante.reset_index(inplace=True)
    visitante = visitante.rename(columns={'index':'VISITOR_TEAM_ID'})
    result = pd.concat([local, visitante], axis=1, sort=False)
    return model.predict_proba(result)
```

```python
partidos=[('WAS','ATL'),('BKN','SAS'),('NYK','OKC'),('BOS','UTA'),('CHI','IND'),('MIN','ORL'),('NOP','MIA'),('DAL','MEM'),('LAL','MIL'),('PHX','POR')]
for i in partidos:
    print(i[0],i[1])
    print(prediceme(i[0],i[1]))
```
