TODOS LOS EQUIPOS DE LOCAL

```python
mediasLocales2019 = df_todos.groupby(['HOME_TEAM_ID']).mean()
mediasLocales2019
```

TODOS LOS EQUIPOS DE VISITANTE

```python
todosVisitantes = games[games['SEASON']>=2019]
```

```python
df_todosVisitantes = todosVisitantes[['VISITOR_TEAM_ID','PTS_away', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away','AST_away', 'REB_away']]
```

```python
mediasVisitantes2019 = df_todosVisitantes.groupby('VISITOR_TEAM_ID').mean()
```

UNIMOS Y LIMPIAMOS DATAFRAMES

```python
df_2019 = games[['GAME_ID','HOME_TEAM_ID', 'VISITOR_TEAM_ID','SEASON','HOME_TEAM_WINS']]
```

ENTRENAMOS EL MODELO

```python
y = df_final.HOME_TEAM_WINS
```

```python
X = df_final.drop(columns =['HOME_TEAM_WINS','SEASON','GAME_ID'])
```

```python
print(y.shape)
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=800, max_depth=20, min_samples_leaf=3, random_state=111, n_jobs = -1)

model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
print('Accurancy', accuracy_score(y_test, y_pred))
print("Precision", precision_score(y_test, y_pred, average='weighted'))
print("Recall", recall_score(y_test, y_pred, average='weighted'))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

y_score = model.predict_proba(X_test)


display(pd.DataFrame({
    "real":y_test,
    "predicted":y_pred,
    "probaLocal":y_score[:,0],
    "probaVisitante":y_score[:,1]
    
}).head(30))
```

```python
teams = pd.read_csv('input/teams.csv')
```

```python
teams = teams[['TEAM_ID', 'ABBREVIATION' ]]
```

```python
teams
```

```python
teams = teams.set_index('ABBREVIATION')
```

```python
def homeTeam (name):
    return teams['TEAM_ID'].loc[name]
```

```python
def awayTeam (name):
    return teams['TEAM_ID'].loc[name]
```

```python
homeTeam('DAL')
```

```python
awayTeam('NOP')
```

```python
def prediceme(local, visitante):
    team1 = pd.DataFrame(mediasLocales2019.loc[homeTeam(local)])
    player_home = pd.DataFrame({homeTeam(local):int(input('Â¿cuantos jugadores del Top5 del equipo local van a jugar? '))
                 }, index=['HOME_TEAM_ID_y'])
    allHomeTeam = pd.concat([team1, player_home], axis=0, sort=False)
    allHomeTeam = allHomeTeam.T
    allHomeTeam = allHomeTeam.reset_index()
    homesTeams = allHomeTeam.rename(columns={'index':'HOME_TEAM_ID'})
    team2 = pd.DataFrame(mediasVisitantes2019.loc[awayTeam(visitante)])
    player_away = pd.DataFrame({awayTeam(visitante):int(input('Â¿cuantos jugadores del Top5 del del equipo visitante van a jugar? '))
                 }, index=['VISITOR_TEAM_ID_y'])
    allAwayTeam = pd.concat([team2, player_away], axis=0, sort=False)
    allAwayTeam = allAwayTeam.T
    allAwayTeam = allAwayTeam.reset_index()
    awayTeams = allAwayTeam.rename(columns={'index':'VISITOR_TEAM_ID'})
    results = pd.concat([homesTeams, awayTeams], axis=1, sort=False)
    return model.predict_proba(results)
```
